```python
doc = nlp(data.iloc[50]['review_text'])
doc
```

```python
for token in doc.noun_chunks:
    print(token)
```

```python
for token in doc:
    if ((token.pos_ == 'NOUN') | (token.pos_ == 'ADJ')) & (token.is_stop == False):
        print(token.text +' | '+token.lemma_+' | '+token.pos_+' | '+str(token.is_stop))
```

```python
def nounadjstring(text):
    try:
        newstring = []
        doc = nlp(text)
        for token in doc:
            if ((token.pos_ == 'NOUN') | (token.pos_ == 'ADJ')) & (token.is_stop == False):
                newstring.append(token.lemma_)
        return ' '.join(str(e) for e in newstring)
    except: 
        return ''
```

```python
nounadjstring(data.iloc[50]['review_text'])
```

```python
data['nounadjtext'] = data['review_text'].map(nounadjstring)
```
